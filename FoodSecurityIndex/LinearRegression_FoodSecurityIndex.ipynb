```python
import pandas as pd
```

```python
# If you are using other folder distribution or drive modify here your path:
path = ''
```

```python
path = path + 'Data_Countries/'
```

```python
values_country = pd.read_excel(path + '2020_GFSI_Data_Table_Data.xlsx', skiprows=5)
values_country
```

```python
values_country.drop(columns = ['Unnamed: 0', 'Unnamed: 2'], inplace = True)
values_country.rename(columns = {'Unnamed: 1' : 'features'},inplace = True)
values_country = values_country[values_country['Average (mean)']  != '.']
values_country.index = values_country.features
values_country.drop(columns = ['Average (mean)', 'Average (median)', 'Minimum', 'Maximum', 'features'], inplace = True)
values_country = values_country.T
values_country
```

```python
scores_country = pd.read_excel(path + '2020_GFSI_Data_Table_Scores.xlsx', skiprows=5)
scores_country
```

```python
scores_country.drop(columns = ['Unnamed: 0', 'Weight', 'Average (mean)', 'Average (median)', 'Minimum', 'Maximum'], inplace = True)
scores_country.rename(columns = {'Unnamed: 1' : 'features'},inplace = True)
#scores_country = scores_country[scores_country['Average (mean)']  != '.']
scores_country.index = scores_country.features
scores_country.drop(columns = ['features'], inplace = True)
scores_country = scores_country.T
scores_country
```

# Linear Regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(scores_country.drop('OVERALL FOOD SECURITY ENVIRONMENT', axis=1), scores_country['OVERALL FOOD SECURITY ENVIRONMENT'])
```
